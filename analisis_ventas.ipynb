```python
import pandas as pd

ventas = pd.read_csv("ventas.csv")
ventas.head()
ventas.info()
ventas.isnull().sum()
ventas.groupby("mes")["monto_total"].sum()
ventas.groupby("categoria")["monto_total"].sum()
import matplotlib.pyplot as plt

ventas.groupby("mes")["monto_total"].sum().plot(kind="bar")
plt.title("Ventas por Mes")
plt.show()

## Conclusiones

## Se identificó que los meses con mayor volumen de ventas fueron Abril y Mayo.
## La categoría Electrónica concentra el mayor monto total de ventas.
## Este análisis permite identificar patrones temporales y categorías clave para la toma de decisiones.

```
